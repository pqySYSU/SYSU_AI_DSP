### 一.掌握几种基本的离散时间信号
1.单位脉冲序列:  
$$ \delta(n)=\left\{
\begin{matrix}
1 , n=0 \\
0 , n\neq0  
\end{matrix}
\right.
$$
2.单位阶跃序列:
$$ u(n)=\left\{
\begin{matrix}
1 , n\geq0 \\
0 , n<0  
\end{matrix}
\right.
$$
3.单频正弦序列:
$$ \left\{
\begin{matrix}
x[n]=A\cos(2\pi fn+\varphi) \\
x[n]=A\sin(2\pi fn+\varphi)
\end{matrix}
\right.
$$
4.单频复指数序列:
$$
x[n]=Ae^{n(j\omega_0+\sigma)}
$$
5.实指数序列:
$$
x[n]=A\alpha ^n
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.sans-serif'] = ['SimHei'] # 使用 SimHei 字体
rcParams['axes.unicode_minus'] = False # 解决坐标轴负号显示问题
```

```python
def task_a(amplitude, frequency, phase, sample_rate, duration):
    """
    生成正弦信号

    参数:
        amplitude: 振幅
        frequency: 频率 (Hz)
        phase: 相位 (弧度)
        sample_rate: 采样率 (Hz)
        duration: 信号时长 (秒)

    返回:
        时间数组，信号数组
    """
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    signal = amplitude * np.sin(2 * np.pi * frequency * t + phase)
    return t, signal
```

```python
def task_b(sample_rate, duration):
    """
    生成单位阶跃信号

    参数:
        sample_rate: 采样率 (Hz)
        duration: 信号时长 (秒)

    返回:
        时间数组，信号数组
    """
    t = np.linspace(-duration / 2, duration / 2, int(sample_rate * duration), endpoint=False)  # 修改时间范围
    signal = np.zeros_like(t)  # 初始化信号数组为全 0
    signal[t >= 0] = 1  # 设置 t >= 0 的部分为 1
    return t, signal
```

```python
def task_c(amplitude, base, frequency, sample_rate, duration):
    """
    生成实指数信号

    参数:
        amplitude: 振幅
        base: 实底数
        frequency: 频率 (Hz)
        sample_rate: 采样率 (Hz)
        duration: 信号时长 (秒)

    返回:
        时间数组，信号数组
    """
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    signal = amplitude * base**(2 * np.pi * frequency * t)  # 修正公式
    return t, signal
```

```python
# 示例用法:
amplitude = 1  # 振幅为 1
frequency = 1  # 频率为 1Hz
phase = 0  # 相位为 0 弧度
sample_rate = 100  # 采样频率为 100Hz
duration = 5  # 信号时长为 5 秒

# 生成正弦信号
t, signal = task_a(amplitude, frequency, phase, sample_rate, duration)

# 绘制图形
plt.figure(figsize=(8, 4))
plt.plot(t, signal)
plt.title('正弦信号 (振幅: {}, 频率: {}Hz, 相位: {}弧度, 采样频率: {}Hz)'.format(
    amplitude, frequency, phase, sample_rate))
plt.xlabel('时间 (秒)')
plt.ylabel('幅值')
plt.grid(True)
plt.show()
```

```python

# 任务 b: 生成单位价格信号 (假设单位价格为 5)
t_b, signal_b = task_b(sample_rate, duration)

plt.plot(t_b, signal_b)
plt.title('任务 b: 单位阶跃信号(采样频率: {}Hz)'.format(sample_rate))
```

```python
# 任务 c: 生成实际数据信号
amplitude = 2
base = 1.2  # 实底数
frequency = 1
sample_rate = 10
duration = 5

t_c, signal_c = task_c(amplitude, base, frequency, sample_rate, duration)


plt.figure(figsize=(8, 4))
plt.plot(t_c, signal_c)
plt.title('实指数信号 (振幅: {}, 底数: {}, 频率: {}Hz, 采样频率: {}Hz)'.format(
    amplitude, base, frequency, sample_rate))
plt.xlabel('时间 (秒)')
plt.ylabel('幅值')
plt.grid(True)
plt.show()
```

```python
def task_d(sample_rate, duration):
    """
    生成单位采样信号

    参数:
        sample_rate: 采样率 (Hz)
        duration: 信号时长 (秒)

    返回:
        时间数组，信号数组
    """
    t = np.arange(0, duration, 1 / sample_rate)  # 生成时间数组
    signal = np.zeros_like(t)  # 初始化信号数组为全 0
    signal[0] = 1  # 在 t=0 处设置为 1
    return t, signal

# 示例用法:
sample_rate = 10  # 采样频率为 10Hz
duration = 5

# 生成单位采样信号
t_d, signal_d = task_d(sample_rate, duration)

# 绘制图形
plt.figure(figsize=(8, 4))
plt.stem(t_d, signal_d, linefmt='b-', markerfmt='bo', basefmt='k-')  # 使用 stem 函数绘制离散信号
plt.title('单位采样信号 (采样频率: {}Hz)'.format(sample_rate))
plt.xlabel('时间 (秒)')
plt.ylabel('幅值')
plt.grid(True)
plt.show()
```

```python
def task_e(amplitude, base, frequency, sample_rate, duration):
    """
    生成复指数信号

    参数:
        amplitude: 振幅
        base: 实底数
        frequency: 频率 (Hz)
        sample_rate: 采样率 (Hz)
        duration: 信号时长 (秒)

    返回:
        时间数组，信号数组
    """
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    signal = amplitude * np.power(base, 2 * np.pi * frequency * t * 1j)  # 使用 1j 表示虚数单位
    return t, signal


# 示例用法:
amplitude = 2
base = 1.5  # 实底数
frequency = 1
sample_rate = 100
duration = 5

# 生成复指数信号
t, signal = task_e(amplitude, base, frequency, sample_rate, duration)


# 绘制实部和虚部
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal.real)
plt.title('复指数信号实部 (振幅: {}, 底数: {}, 频率: {}Hz, 采样频率: {}Hz)'.format(
    amplitude, base, frequency, sample_rate))
plt.xlabel('时间 (秒)')
plt.ylabel('幅值')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(t, signal.imag)
plt.title('复指数信号虚部')
plt.xlabel('时间 (秒)')
plt.ylabel('幅值')
plt.grid(True)

plt.tight_layout()
plt.show()

# 绘制幅值和相位
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, np.abs(signal))
plt.title('复指数信号幅值')
plt.xlabel('时间 (秒)')
plt.ylabel('幅值')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(t, np.angle(signal))
plt.title('复指数信号相位')
plt.xlabel('时间 (秒)')
plt.ylabel('相位 (弧度)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
def generate_gaussian_white_noise(mean, std_dev, num_samples):
    """
    生成高斯白噪声序列

    参数:
        mean: 均值
        std_dev: 标准差
        num_samples: 样本数量

    返回:
        高斯白噪声序列
    """
    noise = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
    return noise


# 示例用法:
mean = 0  # 均值
std_dev = 1  # 标准差
num_samples = 1000  # 样本数量

# 生成高斯白噪声序列
noise = generate_gaussian_white_noise(mean, std_dev, num_samples)

# 绘制图形
plt.figure(figsize=(8, 4))
plt.plot(noise)
plt.title('高斯白噪声序列 (均值: {}, 标准差: {})'.format(mean, std_dev))
plt.xlabel('样本序号')
plt.ylabel('幅值')
plt.grid(True)
plt.show()
```
